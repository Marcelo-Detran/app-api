```python
import win32com.client as win32
import os

# Lista de arquivos CSV e seus destinos em XLSX
arquivos = [
    {
        "csv": r"C:\Users\admin\Desktop\Dash Presidência\Portal Abertas - Presidência.csv",
        "xlsx": r"C:\Users\admin\Desktop\Dash Presidência\Portal Abertas - Presidência.xlsx",
    },
    {
        "csv": r"C:\Users\admin\Desktop\Dash Presidência\Portal Concluídas - Presidência.csv",
        "xlsx": r"C:\Users\admin\Desktop\Dash Presidência\Portal Concluídas - Presidência.xlsx",
    },
    {
        "csv": r"C:\Users\admin\Desktop\Dash Presidência\Presidência - Abertas.csv",
        "xlsx": r"C:\Users\admin\Desktop\Dash Presidência\Presidência - Abertas.xlsx",
    },
    {
        "csv": r"C:\Users\admin\Desktop\Dash Presidência\Presidência - Concluídas.csv",
        "xlsx": r"C:\Users\admin\Desktop\Dash Presidência\Presidência - Concluídas.xlsx",
    }
]

try:
    # Abrir o Excel
    excel_app = win32.Dispatch("Excel.Application")
    excel_app.Visible = False  # Deixa o Excel invisível

    for arquivo in arquivos:
        caminho_csv = arquivo["csv"]
        caminho_xlsx = arquivo["xlsx"]

        if os.path.exists(caminho_csv):
            print(f"Importando '{caminho_csv}'...")
            # Abrir o CSV e salvar como XLSX
            workbook = excel_app.Workbooks.Open(caminho_csv)
            workbook.SaveAs(caminho_xlsx, FileFormat=51)  # Formato 51 = XLSX
            workbook.Close()
            print(f"Arquivo salvo como '{caminho_xlsx}'")
        else:
            print(f"Arquivo CSV não encontrado: {caminho_csv}")

except Exception as e:
    print(f"Ocorreu um erro: {e}")

finally:
    # Fechar o Excel
    excel_app.Quit()
    print("Processo concluído!")
```

```python
import pandas as pd
import os

# Lista de arquivos XLSX a serem combinados
arquivos_xlsx = [
    r"C:\Users\admin\Desktop\Dash Presidência\Portal Abertas - Presidência.xlsx",
    r"C:\Users\admin\Desktop\Dash Presidência\Portal Concluídas - Presidência.xlsx",
    r"C:\Users\admin\Desktop\Dash Presidência\Presidência - Abertas.xlsx",
    r"C:\Users\admin\Desktop\Dash Presidência\Presidência - Concluídas.xlsx",
]

# Caminho do arquivo final combinado
arquivo_final = r"C:\Users\admin\Desktop\Dash Presidência\Extração.xlsx"

# Lista para armazenar os DataFrames
dataframes = []

# Ler cada arquivo XLSX e armazenar em uma lista
for arquivo in arquivos_xlsx:
    if os.path.exists(arquivo):
        try:
            print(f"Lendo arquivo: {arquivo}")
            df = pd.read_excel(arquivo, engine="openpyxl")
            dataframes.append(df)
        except Exception as e:
            print(f"Erro ao ler '{arquivo}': {e}")
    else:
        print(f"Arquivo não encontrado: {arquivo}")

# Concatenar todos os DataFrames em um único DataFrame
if dataframes:
    print("Concatenando arquivos...")
    df_final = pd.concat(dataframes, ignore_index=True)
    
    # Exportar o DataFrame combinado para um novo arquivo Excel
    df_final.to_excel(arquivo_final, index=False, engine="openpyxl")
    print(f"Arquivo final salvo em: {arquivo_final}")
else:
    print("Nenhum arquivo foi processado.")
```

```python
import pandas as pd
from openpyxl import load_workbook
from openpyxl.utils import get_column_letter
from datetime import datetime

# Caminho do arquivo de entrada e saída
arquivo_entrada = r"C:\Users\admin\Desktop\Dash Presidência\Extração.xlsx"
arquivo_saida = r"C:\Users\admin\Desktop\Dash Presidência\Extração_Atualizado.xlsx"

# Carregar o arquivo Excel em um DataFrame
print("Carregando o arquivo...")
df = pd.read_excel(arquivo_entrada, engine="openpyxl")

# Adicionar a coluna com a data de extração (data atual)
data_extracao = datetime.now().strftime("%d/%m/%Y")  # Formato dd/mm/aaaa  # Formato de data
df['Data de Extração'] = data_extracao

# Salvar temporariamente para adicionar fórmulas com openpyxl
df.to_excel(arquivo_saida, index=False, engine="openpyxl")

# Carregar o arquivo salvo com openpyxl
print("Renomeando a aba e adicionando fórmulas no Excel...")
wb = load_workbook(arquivo_saida)

# Renomear a aba 'Sheet1' para 'Planilha1'
if 'Sheet1' in wb.sheetnames:
    wb['Sheet1'].title = 'Planilha1'

# Criar a aba "DE - PARA"
ws_DE_para = wb.create_sheet("DE - PARA")

# Preencher a aba "DE - PARA" com os valores
dados_de_para = [
    ["STATUS PADRONIZADO", "DE PARA"],
    ["BACKLOG - AGUARDANDO PRIORIZAÇÃO", "NOVA (AN)"],
    ["BACKLOG - REVISÃO", "NOVA (AN)"],
    ["BACKLOG - NOVA", "NOVA (AN)"],
    ["NOVA", "BACKLOG (PRODESP)"],
    ["LIBERADA PARA PLANEJAMENTO", "BACKLOG DESENVOLVIMENTO (PRODESP)"],
    ["PLANEJADA", "BACKLOG DESENVOLVIMENTO (PRODESP)"],
    ["EM ANDAMENTO", "DESENVOLVIMENTO (PRODESP)"],
    ["BACKLOG (DESENV. HABILITAÇÃO)", "BACKLOG DESENVOLVIMENTO (PRODESP)"],
    ["HOMOLOGAÇÃO", "HOMOLOGAÇÃO (DS e AN)"],
    ["HOMOLOGADO", "HOMOLOGAÇÃO (DS e AN)"],
    ["AGUARDANDO REAVALIAÇÃO", "HOMOLOGAÇÃO (DS e AN)"],
    ["RESOLVIDA", "AGUARDANDO AVALIAÇÃO (AN)"],
    ["RETORNO DE HOMOLOGAÇÃO", "DESENVOLVIMENTO (PRODESP)"],
    ["EM TESTES", "EM TESTES (DS)"],
    ["TESTE APROVADO", "EM TESTES (DS)"],
    ["CONCLUÍDA (CONCLUIDA)", "CONCLUÍDA"],
    ["PRODUÇÃO", "CONCLUÍDA"],
    ["EXECUÇÃO SUSPENSA", "CANCELADA"],
    ["CANCELADA", "CANCELADA"],
    ["Portal Detran - Solicitações", "Portal Detran"]
]

for row in dados_de_para:
    ws_DE_para.append(row)

# Adicionar fórmula na nova coluna
ultima_coluna = len(df.columns) + 1
nova_coluna_letra = get_column_letter(ultima_coluna)
formula_coluna_nome = "DE - PARA"

# Adicionar cabeçalho da nova coluna com fórmula
ws = wb.active  # Referenciar a aba ativa, que é a Planilha1 após renomear
ws[f"{nova_coluna_letra}1"] = formula_coluna_nome

# Inserir a fórmula em todas as linhas
for linha in range(2, ws.max_row + 1):  # Começa da linha 2 (exclui cabeçalho)
    ws[f"{nova_coluna_letra}{linha}"] = "=SE(D{0}='Nova','DE - PARA'!$B$5,SE(D{0}='Em Andamento','DE - PARA'!$B$8,SE(D{0}='Aguardando Reavaliação','DE - PARA'!$B$12,SE(D{0}='Resolvida','DE - PARA'!$B$13,SE(D{0}='Concluída','DE - PARA'!$B$17,SE(D{0}='Cancelada','DE - PARA'!$B$21,' ')))))".format(linha)

# Salvar o arquivo final
wb.save(arquivo_saida)
print(f"Arquivo atualizado salvo em: {arquivo_saida}")
```
