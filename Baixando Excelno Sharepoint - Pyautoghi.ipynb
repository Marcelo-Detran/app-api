```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import pyautogui
import time
import logging
import os
import shutil

# Configuração do WebDriver com a pasta de download
download_dir = r"C:\Users\admin\Downloads"
destino_folder = r"C:\Users\admin\Desktop\Dash Presidência"

# Configuração de Logs
logging.basicConfig(level=logging.INFO, format="%(asctime)s - %(message)s")

# Configurações do SharePoint
site_url = "https://governosp.sharepoint.com/teams/DETRAN-gestao.sistemas"
arquivo_path = "Gestao De Dados e Dashboads/Evolutiva/Dash_Sustentação/extração sustentação.xlsx"
download_folder = r"C:\\Users\\admin\\Desktop\\Dash Presidência"
file_name = "extração sustentação.xlsx"

# Configuração do Chrome WebDriver
chrome_options = Options()
chrome_options.add_argument("--start-maximized")
chrome_options.add_argument("--disable-blink-features=AutomationControlled")
chrome_options.add_experimental_option("prefs", {
    "download.default_directory": download_folder,
    "download.prompt_for_download": False,
    "download.directory_upgrade": True,
    "safebrowsing.enabled": True
})

driver = webdriver.Chrome(options=chrome_options)

def login_e_baixar_arquivo():
    try:
        logging.info("Acessando o SharePoint...")
        driver.get(site_url)

        # Login no SharePoint
        logging.info("Fazendo login...")
        WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.ID, "i0116"))).send_keys("marcelo.cota@apoioprodesp.sp.gov.br")
        driver.find_element(By.ID, "idSIButton9").click()

        WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.ID, "i0118"))).send_keys("@15MN09cp77@")
        driver.find_element(By.ID, "idSIButton9").click()

        # Confirmar a opção "Permaneça logado"
        try:
            WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.ID, "idSIButton9"))).click()
        except:
            pass

        # Pesquisa pelo arquivo no SharePoint
        logging.info("Pesquisando pelo arquivo...")
        search_box = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, "//input[@aria-label='Pesquisar neste site']")))
        search_box.send_keys(arquivo_path)
        search_box.send_keys(Keys.RETURN)

        # Clique no link do arquivo
        logging.info("Localizando o arquivo...")
        file_link = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.PARTIAL_LINK_TEXT, file_name)))
        file_link.click()

        logging.info(f"Arquivo {file_name} baixado com sucesso.")
    
    except Exception as e:
        logging.error(f"Erro: {e}")
    finally:
        driver.quit()

def confirmar_download():
    try:
        logging.info("Confirmando o download com PyAutoGUI...")
        time.sleep(5)  # Tempo para a janela de diálogo aparecer

        # Tenta localizar e clicar no botão "Salvar"
        salvar_btn = pyautogui.locateOnScreen('salvar.png', confidence=0.9)
        if salvar_btn:
            pyautogui.click(salvar_btn)
            logging.info("Botão 'Salvar' encontrado e clicado.")
        else:
            logging.warning("Botão 'Salvar' não encontrado. Tentando pressionar 'Enter'.")
            pyautogui.press("enter")

    except Exception as e:
        logging.error(f"Erro ao confirmar o download: {e}")

# Executa as funções
login_e_baixar_arquivo()
confirmar_download()
```
