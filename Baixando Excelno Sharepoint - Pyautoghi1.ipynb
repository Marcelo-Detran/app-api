```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import pyautogui
import time
import logging
import os
import shutil

# Configuração de diretórios
download_dir = r"C:\Users\admin\Downloads"
destino_folder = r"C:\Users\admin\Desktop\Dash Presidência"
arquivo_nome = "extração sustentaçao.xlsx"  # Nome parcial do arquivo esperado

# Configuração de Logs
logging.basicConfig(level=logging.INFO, format="%(asctime)s - %(message)s")

# URL do SharePoint
site_url = "https://governosp.sharepoint.com/teams/DETRAN-gestao.sistemas"

# Configuração do Chrome WebDriver
chrome_options = Options()
chrome_options.add_argument("--start-maximized")  # Maximiza a janela do navegador
chrome_options.add_experimental_option("prefs", {
    "download.default_directory": download_dir,
    "download.prompt_for_download": False,
    "safebrowsing.enabled": True
})
driver = webdriver.Chrome(options=chrome_options)

def acessar_pagina_login():
    logging.info("Acessando a página de login do SharePoint...")
    driver.get(site_url)

def mover_arquivo():
    try:
        logging.info("Verificando e movendo o arquivo baixado...")

        # Aguarda alguns segundos para garantir que o download foi concluído
        time.sleep(30)

        # Lista os arquivos na pasta de download
        arquivos_na_pasta = os.listdir(download_dir)
        logging.info(f"Arquivos encontrados: {arquivos_na_pasta}")  # Log dos arquivos encontrados

        # Procura o arquivo que contém o nome específico
        for arquivo in arquivos_na_pasta:
            if arquivo_nome in arquivo:  # Verifica se o nome parcial está no arquivo
                origem_arquivo = os.path.join(download_dir, arquivo)
                destino_arquivo = os.path.join(destino_folder, arquivo)

                # Verifica se o arquivo tem tamanho maior que 0 e se não está sendo transferido
                while True:
                    tamanho_arquivo_antes = os.path.getsize(origem_arquivo)
                    time.sleep(1)  # Aguarda 1 segundo
                    tamanho_arquivo_depois = os.path.getsize(origem_arquivo)
                    
                    # Se o tamanho não mudou, o arquivo foi totalmente baixado
                    if tamanho_arquivo_antes == tamanho_arquivo_depois:
                        break

                # Move o arquivo para a pasta de destino
                shutil.move(origem_arquivo, destino_arquivo)
                logging.info(f"Arquivo '{arquivo}' movido para: {destino_folder}")
                return  # Encerra após mover o arquivo
        
        logging.warning("Arquivo não encontrado na pasta de downloads!")

    except Exception as e:
        logging.error(f"Erro ao mover o arquivo: {e}")


def interagir_com_interface():
    """ Interage com a interface usando o PyAutoGUI """
    pyautogui.PAUSE = 1.0
    time.sleep(5)

    # Exemplo de cliques (ajuste conforme necessário)
    pyautogui.moveTo(x=951, y=813, duration=1)
    pyautogui.click(x=951, y=813)

    time.sleep(8)
    pyautogui.moveTo(x=993, y=537, duration=1)
    pyautogui.click(x=993, y=537)

    time.sleep(8)
    pyautogui.moveTo(x=1068, y=503, duration=1)
    pyautogui.click(x=1068, y=503)

    time.sleep(8)
    pyautogui.write("15072016")

    time.sleep(8)
    pyautogui.moveTo(x=231, y=579, duration=1)
    pyautogui.click(x=231, y=579)

    time.sleep(8)
    pyautogui.moveTo(x=525, y=573, duration=1)
    pyautogui.click(x=525, y=573)
    
    time.sleep(8)
    pyautogui.moveTo(x=528, y=573, duration=1)
    pyautogui.click(x=528, y=573)

    time.sleep(8)
    pyautogui.moveTo(x=834, y=626, duration=1)
    pyautogui.click(x=834, y=626)

    time.sleep(8)
    pyautogui.moveTo(x=908, y=670, duration=1)
    pyautogui.click(x=908, y=670)

# Fluxo principal
try:
    acessar_pagina_login()
    interagir_com_interface()
    mover_arquivo()
finally:
    driver.quit()
    logging.info("Processo concluído.")
```
